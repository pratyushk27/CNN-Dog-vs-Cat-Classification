```python
# Preprocessing the Training set
train_datagen = ImageDataGenerator(
    rescale=1/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
# Preprocessing the Test set
test_datagen = ImageDataGenerator(
    rescale=1/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
base_dir = '/content/cat&dog'

# Create datasets
train_datagen = image_dataset_from_directory(base_dir,
												image_size=(200,200),
												subset='training',
												seed = 1,
												validation_split=0.1,
												batch_size= 32)
test_datagen = image_dataset_from_directory(base_dir,
												image_size=(200,200),
												subset='validation',
												seed = 1,
												validation_split=0.1,
												batch_size= 32)


```

Building the CNN model

```python
# Building the CNN

model = tf.keras.models.Sequential([
    # First convolutional layer
    tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=[200, 200, 3]),
    tf.keras.layers.MaxPool2D(2),

    # Second convolutional layer
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPool2D(2),

    # Flattening
    tf.keras.layers.Flatten(),

    # Full Connection
    tf.keras.layers.Dense(128, activation='relu'),

    # Output Layer
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()                    # Model summary
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])            # Compile the model
```

```python
# Training the CNN on the Training set and evaluating it on the Test set
```

```python
history = model.fit(train_datagen,
          epochs=10,
          validation_data=test_datagen)
```

Predict the model

```python
from google.colab.patches import cv2_imshow
```

```python
input_image_path = input('Path of the image to be predicted: ')

input_image = cv2.imread(input_image_path)

cv2_imshow(input_image)

input_image_resize = cv2.resize(input_image, (200,200))

scaled_image = input_image_resize/255

image_reshaped = np.reshape(scaled_image, [1,200, 200, 3])

input_pred = model.predict(image_reshaped)

print(input_pred)

input_pred_label = np.argmax(input_pred)

print(input_pred_label)

if input_pred_label == 0:
  print('The image is of a Cat')

else:
  print('The image is of a Dog')
```
